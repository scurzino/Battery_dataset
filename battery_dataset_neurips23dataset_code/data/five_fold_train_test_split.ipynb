```python
import torch

# Controlliamo se abbiamo GIA' disattivato il blocco
if not hasattr(torch, '_blocco_disattivato'):
    
    _original_load = torch.load
    
    def _unsafe_load(*args, **kwargs):
        kwargs['weights_only'] = False
        return _original_load(*args, **kwargs)
        
    torch.load = _unsafe_load
    
    # Mettiamo un'etichetta per ricordarci che l'abbiamo fatto
    torch._blocco_disattivato = True 
    print("✅ Blocco PyTorch disattivato con successo. Procedere!")
else:
    print("⚡ Il blocco era già disattivato, nessun rischio di loop. Procedere!")
```

```python
import numpy as np
import os
import pickle
from glob import glob
import torch
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
import random
from tqdm import tqdm
```

```python
data_dir = "."
```

# generate new all_car_dict based on battery_dataset1, battery_dataset2 and battery_dataset3

# battery_dataset1

```python
data_path = data_dir+'/battery_dataset1/data'

data_pkl_files = glob(data_path+'/*.pkl')

ind_pkl_files = []
ood_pkl_files = []
car_num_list = []

ood_car_num_list1 = set()
ind_car_num_list1 = set()

all_car_dict = {}

for each_path in tqdm(data_pkl_files):
#     print(each_path)
    this_pkl_file = torch.load(each_path)
    this_car_number = this_pkl_file[1]['car']
    if this_pkl_file[1]['label'] == '00':
        ind_pkl_files.append(each_path)
        ind_car_num_list1.add(this_car_number)
    else:
        ood_pkl_files.append(each_path)
        ood_car_num_list1.add(this_car_number)
    car_num_list.append(this_pkl_file[1]['car'])
    if this_car_number not in all_car_dict:
        all_car_dict[this_car_number] = []
        all_car_dict[this_car_number].append(each_path)
    else:
        all_car_dict[this_car_number].append(each_path)
```

```python
# shuffle
random.seed(0)
ind_sorted = sorted(ind_car_num_list1)
random.shuffle(ind_sorted)
print(ind_sorted)
ood_sorted = sorted(ood_car_num_list1)
random.shuffle(ood_sorted)
print(ood_sorted)        
ind_odd_dict = {}
ind_odd_dict["ind_sorted"],ind_odd_dict["ood_sorted"] = ind_sorted, ood_sorted
os.makedirs('../five_fold_utils', exist_ok=True)
np.save('../five_fold_utils/ind_odd_dict1.npz', ind_odd_dict)
```
